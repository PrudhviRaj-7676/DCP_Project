```python
import pandas as pd
import pickle
df = pd.read_csv('selected_hospitals.csv')
df.head()
```

```python
df['payer'] = df['payer'].str.strip()
df['code'] = df['code'].str.strip()

# Drop rows where 'payer' has unwanted values
unwanted_payers = ['GROSS CHARGE', 'CASH PRICE', 'MAX', 'MIN']
df_filtered = df[~df['payer'].isin(unwanted_payers)]

# Drop rows where 'code' has the value 'NONE'
df_filtered = df_filtered[df_filtered['code'] != 'NONE']

# Display the first few rows to verify the changes
df_filtered.head()
```

```python
df_filtered['gross_charges'] = df_filtered['price'] * 1.08  # Gross charges increased by 8%
df_filtered['maximum_price'] = df_filtered['price'] * 1.05  # Maximum price increased by 5%
df_filtered.head()
```
